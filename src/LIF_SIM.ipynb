#### Check R Parameter


```python
R_values = [1, 10, 20]

for R in R_values:
    sim3 = BaseSimulator(0.001)

    current = StepFunction(value=3000, t0=0.5)

    neuron_model = LIF(
        tau=10,
        u_rest=-65,
        u_reset=-70,
        threshold=-40,
        R=R,
        v_init_mode="normal(-40, 20)",
    )

    sim3.buildNeuronGroup(current, neuron_model)

    sim3.exec(15000)

    sim3.plot_just_voltage()
    # sim3.plot_spike_times()
```

#### Check $\tau$ Parameter


```python
tau_values = [1, 10, 20]

for tau in tau_values:
    sim3 = BaseSimulator(0.001)

    current = StepFunction(value=2000, t0=0.5)

    neuron_model = LIF(
        tau=tau,
        u_rest=-65,
        u_reset=-70,
        threshold=-40,
        R=3,
        v_init_mode="normal(-40, 20)",
    )

    sim3.buildNeuronGroup(current, neuron_model)

    sim3.exec(15000)

    sim3.plot_just_voltage()
```

#### F-I Curve


##### Without Noise


```python
y, x = [], []
for i in range(9):
    current_value = (i+1) * 200
    sim1 = BaseSimulator(0.01)

    current = ConstantCurrent(value=current_value)

    neuron_model = LIF(
        tau=10,
        u_rest=-65,
        u_reset=-70,
        threshold=-40,
        R=10,
        v_init_mode="normal(-40, 20)",
    )

    sim1.buildNeuronGroup(current, neuron_model)

    sim1.exec(100)

    y.append(sim1.net['spike_counts', 0][-1].item() / 100)
    x.append(current_value)
```

```python
plt.figure(figsize=(4.5, 3.5))
plt.plot(x, y)
plt.xlabel("current")
plt.ylabel("frequency")
plt.title("F-I curve")
plt.show()
```

```python
y, x = [], []
for i in range(9):
    current_value = (i+1) * 200
    sim1 = BaseSimulator(0.01)

    current = ConstantCurrent(
        value=current_value, with_noise=True, noise_std=5, noise_level=30)

    neuron_model = LIF(
        tau=10,
        u_rest=-65,
        u_reset=-70,
        threshold=-40,
        R=10,
        v_init_mode="normal(-40, 20)",
    )

    sim1.buildNeuronGroup(current, neuron_model)

    sim1.exec(100)

    y.append(sim1.net['spike_counts', 0][-1].item() / 100)
    x.append(current_value)
```
