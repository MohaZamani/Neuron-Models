#### Consatnt Current


```python
sim1 = BaseSimulator(0.001)

current = ConstantCurrent(value=1000)

neuron_model = AELIF(
    tau=10,
    u_rest=-65,
    u_reset=-70,
    threshold=0,
    threshold_rh=-35,
    R=3,
    delta_t=1,
    v_init_mode="normal(-50, 14)",
    w_init_mode="normal(5, 10)",
    a_k=-20,
    b_k=45,
    tau_k=95
)

sim1.buildNeuronGroup(current, neuron_model)

sim1.exec(3000)

sim1.plot_voltage_current()
sim1.plot_spike_times()
```

#### F-I Curve


##### Without Noise


```python
y, x = [], []
for i in range(20):
    current_value = (i+1) * 200
    sim1 = BaseSimulator(0.01)

    current = ConstantCurrent(value=current_value)

    neuron_model = AELIF(
        tau=10,
        u_rest=-65,
        u_reset=-70,
        threshold=0,
        threshold_rh=-35,
        R=10,
        delta_t=1,
        v_init_mode="normal(-50, 14)",
        w_init_mode="normal(5, 10)",
        a_k=-20,
        b_k=45,
        tau_k=95
    )

    sim1.buildNeuronGroup(current, neuron_model)

    sim1.exec(100)

    y.append(sim1.net['spike_counts', 0][-1].item() / 100)
    x.append(current_value)
```

```python
plt.figure(figsize=(4.5, 3.5))
plt.plot(x, y)
plt.xlabel("current")
plt.ylabel("frequency")
plt.title("F-I curve")
plt.show()
```

##### With Noise


```python
y, x = [], []
for i in range(20):
    current_value = (i+1) * 200
    sim1 = BaseSimulator(0.01)

    current = ConstantCurrent(
        value=current_value, with_noise=True, noise_std=5, noise_level=50)

    neuron_model = AELIF(
        tau=10,
        u_rest=-65,
        u_reset=-70,
        threshold=0,
        threshold_rh=-35,
        R=10,
        delta_t=1,
        v_init_mode="normal(-50, 14)",
        w_init_mode="normal(5, 10)",
        a_k=-20,
        b_k=45,
        tau_k=95
    )

    sim1.buildNeuronGroup(current, neuron_model)

    sim1.exec(100)

    y.append(sim1.net['spike_counts', 0][-1].item() / 100)
    x.append(current_value)
```

```python
plt.figure(figsize=(4.5, 3.5))
plt.plot(x, y)
plt.xlabel("current")
plt.ylabel("frequency")
plt.title("F-I curve")
plt.show()
```
